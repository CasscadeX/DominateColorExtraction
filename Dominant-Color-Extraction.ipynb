```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.cluster import KMeans
```

```python
img = cv2.imread("giraffe.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
def K_Means(img,n):
    k = n
    km = KMeans(n_clusters=k)
    h = img.shape[0]
    w = img.shape[1]
    pixels = img.reshape((-1,3))
    km.fit(pixels)
    colors = km.cluster_centers_
    colors = colors.round()
    colors = np.array(colors,dtype="uint")
    for c in colors:
        a = np.zeros((100,100,3),dtype="uint")
        a[:,:,:] = c
    new_img = np.zeros((h*w,3),dtype="uint")
    for ix in range(new_img.shape[0]):
        new_img[ix] = colors[km.labels_[ix]]
    new_img = new_img.reshape((h, w, 3))
    return new_img
    
```

```python
images = []
```

```python
image = K_Means(img,20)
images.append(image)
plt.imshow(image)
```

```python
images = np.array(images)
```
